```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

**1. Calculate the mean of each feature.**

```python
mean = df.mean()
print('Mean:\n', mean)
```

 **2. Calculate the median of each feature.**

```python
median = df.median()
print('Median:\n', median)
```

**3. Calculate the mode of each feature.**

```python
mode = df.mode().iloc[0]
print('Mode:\n', mode)
```

**4. Calculate the standard deviation of each feature.**

```python
std_dev = df.std()
print('Standard Deviation:\n', std_dev)
```

**5. Calculate the variance of each feature.**

```python
variance = df.var()
print('Variance:\n', variance)
```

**6. Calculate the range of each feature.**

```python
range_val = df.max() - df.min()
print('Range:\n', range_val)
```

**7. Calculate the interquartile range (IQR) of each feature**

```python
iqr = df.quantile(0.75) - df.quantile(0.25)
print('Interquartile Range:\n', iqr)
```

**8. Plot histograms for each feature**

```python
for col in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()
```

**9. Plot boxplots for each feature.**

```python
for col in df.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

**10. Plot pair plots to visualize relationships between features.**

```python
sns.pairplot(df)
plt.show()
```
